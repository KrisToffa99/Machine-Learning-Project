```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
epl = pd.read_csv("EPL.csv")
plt.figure(figsize=(15,5))

plt.plot(epl.Year, epl.Chelsea, label ='Chelsea', marker ='.')
plt.plot(epl.Year, epl.Arsenal, label ='Arsenal', marker ='.')
plt.plot(epl.Year, epl.ManU, label = 'Man. United', marker ='.')
plt.plot(epl.Year, epl.ManCity,label = 'ManCity', marker ='.')
plt.plot(epl.Year, epl.Liverpool, label ='Liverpool', marker ='.')
plt.plot(epl.Year, epl.Tottenham, label = 'Tottenham', marker ='.')

plt.title('EPL Top Clubs Ratings Since 2010',  fontdict={'fontweight':'bold', 'fontsize':20})

plt.xticks(epl.Year[ : ])
plt.yticks(np.arange(0, 11, 1)) 

plt.xlabel('Year', fontdict={'fontweight':'bold', 'fontsize':20})
plt.ylabel('Position', fontdict={'fontweight':'bold', 'fontsize':20})

plt.legend()

plt.show()
```
